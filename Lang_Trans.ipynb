```python
from googletrans import Translator, LANGUAGES # type: ignore
translator = Translator()

lang_name_to_code = {v: k for k, v in LANGUAGES.items()}

def errors():
    print('Invalid language code.')
    print("Available Languages: \n")
    print(f'Language Codes: {lang_name_to_code.keys()}')
    print(f'language_names: {lang_name_to_code.values()}')
```

```python
print(lang_name_to_code.keys())
```

```python
print(lang_name_to_code.values())
```

```python
translated_from = input('Enter the language you want to translate from: ').lower().strip()
text = input('Enter the text you want to translate: ').strip()
translated_to = input('Enter the language you want to translate to: ').lower().strip()

if translated_from in lang_name_to_code and translated_to in lang_name_to_code:
    try:
        src_code = lang_name_to_code[translated_from]
        dest_code = lang_name_to_code[translated_to]

        translation = await translator.translate(
            text,
            src=src_code,
            dest=dest_code
        )
        print("Translated text: ")
        print(translation.text)
    except Exception as e:
        print('Error: ',e)
else:
    errors()
```
