```python
import speech_recognition as speech # type: ignore

recognizer = speech.Recognizer()
codes = ['en-IN(English (India))','en-US(English (US))','hi-IN(Hindi)','te-IN(Telugu)','ur-PK(Urdu)','de-DE(German)','ar-SA(Arabic)','es-ES(Spanish)']
lang = input("Enter language code from the below: ")
print(codes)

with speech.Microphone() as source:
        print("Listening....")
        audio = recognizer.listen(source)
    
try:
    text = recognizer.recognize_google(audio,language=lang)
    print("You said: ",text)
except speech.UnknownValueError:
    print("Couldn't understand audio")
except speech.RequestError:
    print("Could not connect to speech service")
```
